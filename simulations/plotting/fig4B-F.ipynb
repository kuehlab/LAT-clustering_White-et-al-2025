```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from glob import glob
import h5py

types = ['WT', 'Y132F', 'G131D']
cmap = plt.get_cmap('tab10')
colors = [cmap(0), cmap(7), [0.886,0.6745,0.2745]]
strength = [1, 0.3, 0.1, 0.03, 0.01, 0.003, 0.001]
```

```python
base_dir = '../run_sims/kinetic_proofreading/'
```

## bound TCRs

```python
num_TCR = pd.read_csv(f'{base_dir}kp_n_bound_TCR.csv',index_col=0)

plt.figure(figsize=[18,1.5])
for i,s in enumerate(strength):
    plt.subplot(1,7,7-i)
    
    plt.plot(num_TCR['time'], num_TCR[f'type0_str{s}_mean'],linewidth=3)
#     plt.fill_between(num_TCR['time'],num_TCR[f'type0_str{s}_low'],num_TCR[f'type0_str{s}_high'],alpha=0.5)
    
    if i != 6 :
        plt.yticks([])
    else:
        plt.yticks([0,1,2],fontsize=18)
    plt.xticks([0,50,100],fontsize=18)
    plt.ylim([0,2])
plt.savefig('KP_num_bound_TCR.png',dpi=1000,bbox_inches='tight')
```

## num clustered vs time

```python
nclust = pd.read_csv(f'{base_dir}kp_max_clustered_time.csv',index_col=0)

plt.figure(figsize=[18,1.5])
for i,s in enumerate(strength):
    plt.subplot(1,7,7-i)
    
    plt.plot(nclust['time'], nclust[f'type0_str{s}_mean'],linewidth=3)
#     plt.fill_between(num_TCR['time'],nclust[f'type0_str{s}_low'],nclust[f'type0_str{s}_high'],alpha=0.5)
    
    if i != 6 :
        plt.yticks([])
    else:
        plt.yticks([0,10,20],fontsize=18)
    plt.xticks(fontsize=18)
    plt.ylim([0,23])
plt.savefig('KP_num_in_clusters.png',dpi=1000,bbox_inches='tight')
```

## fraction with clusters vs strength

```python
largest_clust = pd.read_csv(f'{base_dir}kp_largest_cluster.csv',index_col=0)

frac_clust = {}
for i in range(3):
    fracs = []
    for ii,s in enumerate(strength):
        fracs.append(np.mean(largest_clust[f'type{i}_str{s}']>50))
    frac_clust[f'type{i}'] = fracs

plt.figure(figsize=[10,6])
for i in range(3):
    plt.plot(np.array(strength)*50,frac_clust[f'type{i}'],'o-',
             color=colors[i],markersize=10,linewidth=3,label=types[i])
    
plt.semilogx()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

leg = plt.legend(loc=[1.1,0.3],fontsize=24)

for line in leg.get_lines():
    line.set_linewidth(4.0)

plt.savefig('KP_frac_sims_with_cluster.png',dpi=1000,bbox_inches='tight')
```

## cluster size vs strength

```python
clust_sizes = pd.read_csv(f'{base_dir}kp_max_clustered_distr.csv',index_col=0)

plt.figure(figsize=[10,6])
for i in [1,2,0]:
    
    mean_clust_size = np.zeros([7])
    clust_size_err = np.zeros([2,7])
    for ii,s in enumerate(strength):
        mean_clust_size[ii] = np.mean(clust_sizes[f'type{i}_str{s}'])
        
        clust_size_err[0,ii] = np.quantile(clust_sizes[f'type{i}_str{s}'],0.25)
        clust_size_err[1,ii] = np.quantile(clust_sizes[f'type{i}_str{s}'],0.75)
        
    plt.errorbar(np.array(strength)*50,mean_clust_size,yerr=np.abs(clust_size_err-mean_clust_size),
                 capsize=6,linewidth=3,markersize=10,elinewidth=3,capthick=3,color=colors[i],
                 fmt='o-',label=types[i],barsabove=True)
    
plt.semilogx()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

leg = plt.legend(loc=[1.1,0.3],fontsize=24,ncol=3)

for line in leg.get_lines():
    line.set_linewidth(4.0)

plt.savefig('KP_max_cluster_size.png',dpi=1000,bbox_inches='tight')
```
