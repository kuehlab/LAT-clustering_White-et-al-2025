```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from scipy.signal import savgol_filter
from glob import glob
import h5py
from matplotlib.ticker import FixedLocator

types = ['WT', 'Y132F', 'G131D']
cmap = plt.get_cmap('tab10')
colors = [cmap(0), cmap(7), [0.886,0.6745,0.2745]]
```

```python
base_dir = '../run_sims/single_TCR/'
```

```python
def get_init_time(size_distr,size_min):

    has_cluster = np.sum(size_distr[:,:,size_min:],axis=2).astype(bool)
    cluster_locs = np.where(has_cluster)
    init_time = np.zeros([size_distr.shape[0]])

    for t in range(size_distr.shape[0]):
        if np.sum(cluster_locs[0]==t):
            init_time[t] = cluster_locs[1][cluster_locs[0]==t][0]
        else:
            init_time[t] = np.nan
    
    return init_time

def get_times_and_sizes(clust_sizes,slopes,min_slope,min_duration):

    growing = slopes >= min_slope
    times = np.zeros(clust_sizes.shape[0])*np.nan
    sizes = np.zeros(clust_sizes.shape[0])*np.nan
    
    for j in range(clust_sizes.shape[0]):
        consecutive = 0
        for t in range(clust_sizes.shape[1]-1):
            if growing[j,t]:
                consecutive += 1
            else:
                consecutive = 0
                
            if consecutive >= min_duration:
                times[j] = (t-consecutive+1)/10
                sizes[j] = clust_sizes[j,(t-consecutive+1)]
                break
    return times, sizes
```

```python
data = h5py.File(f'{base_dir}type0_cluster_sizes.h5','r')['sizes'][:]
data.shape
```

```python
cluster_size = np.tile(np.array(range(1,1501)).reshape([1,1500]),[1201,1])
where_cluster = data > 0
max_cluster_sizes = np.zeros([1500,1201])
for j in range(data.shape[0]):
    sim_clusters = where_cluster[j,:,:]
    sim_clusters = cluster_size*sim_clusters
    max_cluster_sizes[j,:] = np.max(sim_clusters,axis=1)

smoothed_max_clust = savgol_filter(max_cluster_sizes,window_length=11,polyorder=1,axis=1)
max_clust_slope = smoothed_max_clust[:,1:] - smoothed_max_clust[:,:-1]
```

## changing the definition of a cluster

```python
cmap = plt.get_cmap('viridis')

fig = plt.figure(figsize=[8,8])

n_per_bin = 20
for i,s in enumerate([5,10,20,30,40,50]):

    ax = plt.subplot(6,1,i+1)

    init_time = get_init_time(data,s)
    init_time = init_time[~np.isnan(init_time)]/10

    print(s, len(init_time)/1500)

    bin_min = 0# np.min(init_time)
    bin_max = np.max(init_time)
    n_bins = int(np.sum(len(init_time))/(n_per_bin+1))
    bin_width = (bin_max-bin_min)/(n_bins)

    bin_min = bin_min - 1.1*bin_width
    bin_max = bin_max + 1.1*bin_width

    n = 15
    e = 3
    w = (120**(1/e))/n
    l = 120**(1/e) + w
    sns.histplot(init_time,label=f'{s}',# ({round(len(init_time)/15,1)}%)',
                 stat='density',element='step',fill=None,color=cmap((6-i)/6),linewidth=2,
                 bins=[-1,0,1.5]+list(np.linspace(3**(1/e),l,n-1)**e))#np.linspace(bin_min,bin_max,n_bins+2))

    plt.xlim(-5,125)
    # plt.xlim([-0.5,40])
    # plt.yticks([0,0.02,0.04,0.06],fontsize=24)
    plt.yticks([])
    plt.xticks([])
    plt.ylabel('')

    if i==5:
        plt.xticks([0,20,40,60,80,100,120],fontsize=24)
        # plt.xticks([0,10,20,30,40],fontsize=24)

fig.subplots_adjust(wspace=0, hspace=0)

# leg = plt.legend(loc=[1.1,0],fontsize=24)
# for legobj in leg.get_lines():
#     legobj.set_linewidth(3.0)

plt.savefig(f'init_time_distr_for_size_range.png',dpi=1000,bbox_inches='tight')
```
