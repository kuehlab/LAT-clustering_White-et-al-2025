```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from glob import glob
import h5py

types = ['WT', 'WT_high-res', 'Y132F', 'G131D']
cmap = plt.get_cmap('tab10')
colors = [cmap(0), cmap(7), [0.886,0.6745,0.2745]]
```

```python
base_dir = '../run_sims/single_TCR/'
```

## max size distribution

```python
plt.figure(figsize=[8,8])
sizes_sim = pd.read_csv(f'{base_dir}single_TCR_max_clustered_distr.csv',index_col=0)
sizes_real = pd.read_csv('McAffee_data/2C_cluster_sizes.csv')

sns.histplot(sizes_sim[sizes_sim['type0']>50]['type0'],
             label='Sim',stat='density',element='step',bins=np.linspace(50,800,25),alpha=0.8)
sns.histplot(sizes_real['n_lat *'],label='Exp',stat='density',
             element='step',fill=False,linewidth=3,color='k',bins=np.linspace(50,800,25))
plt.legend(loc=[1.1,0.7],fontsize=24)
plt.xlabel('Cluster Size')

plt.xlim([50,750])
plt.xticks(fontsize=24)
plt.yticks([0,0.003,0.006,0.009,0.012],[0,3,6,9,12],fontsize=24)

plt.xlabel('')
plt.ylabel('')

plt.savefig('type0_max_num_clust_distr_full.png',dpi=1000,bbox_inches='tight')
```

```python
plt.figure(figsize=[8,8])
sizes_sim = pd.read_csv(f'{base_dir}single_TCR_max_clustered_distr.csv',index_col=0)
sizes_real = pd.read_csv('McAffee_data/2C_cluster_sizes.csv')

sim_bars,bins = np.histogram(sizes_sim['type0'],bins=np.linspace(50,800,25))
exp_bars,_ = np.histogram(sizes_real['n_lat *'],bins=np.linspace(50,800,25))

#scale by peak of experimental distribution
exp_bars = exp_bars/exp_bars[6]
sim_bars = sim_bars/sim_bars[6]

w = 750/24
x = (bins[:-1]+bins[1:])/2
plt.bar(x,sim_bars,width=w,label='Sim',alpha=0.8)
plt.step(bins[1:],exp_bars,label='Exp',color='k',linewidth=3)
# sns.histplot(sizes_sim[sizes_sim['type0']>50]['type0'],
#              label='Sim',stat='density',element='step',bins=np.linspace(50,800,25))
# sns.histplot(sizes_real['n_lat *'],label='Exp',stat='density',
#              element='step',fill=False,linewidth=3,color='k',bins=np.linspace(50,800,25))
plt.legend(loc=[1.1,0.7],fontsize=24)
plt.xlabel('Cluster Size')

plt.xlim([bins[4],750])
plt.ylim([0,3.1])

plt.xticks(fontsize=24)
plt.yticks([0,0.8,1.6,2.4],fontsize=24)

plt.xlabel('')
plt.ylabel('')

plt.savefig('type0_max_num_clust_distr_zoom.png',dpi=1000,bbox_inches='tight')
```

## init time distr

```python
plt.figure(figsize=[8,8])
times_sim = pd.read_csv(f'{base_dir}single_TCR_init_time.csv',index_col=0)
times_real = pd.read_csv('McAffee_data/4B_wt_delays.csv')

sns.histplot(times_sim['type0']/10,
             label='Sim',stat='density',element='step',bins=np.linspace(-6,126,25),alpha=0.8)
sns.histplot(times_real['# LAT delay time (s)'],label='Exp',stat='density',
             element='step',fill=False,linewidth=3,color='k',bins=np.linspace(-8,128,17))

plt.legend(loc=[1.1,0.7],fontsize=24)
plt.xlabel('Delay Time')

plt.xlim([-4,124])
plt.xticks(fontsize=24)
plt.yticks([0,0.005,0.01,0.015,0.02,0.025,0.03,0.035],[0,5,10,15,20,25,30,35],fontsize=24)

plt.xlabel('')
plt.ylabel('')
    
plt.savefig(f'wt_init_time_distr.png',dpi=1000,bbox_inches='tight')
```

## frac with clusters

```python
clust_sizes = pd.read_csv(f'{base_dir}single_TCR_largest_cluster.csv',index_col=0)
real_frac = pd.read_csv('McAffee_data/3D_clust_prob.csv')

bars = np.zeros([4])
errs = np.zeros([4])
for i,t in enumerate([0,2,3]):
    bars[i+1] = np.mean(clust_sizes[f'type{t}']>50)
    
bars[0] = real_frac[' p_lat mcc'].values[-1]
errs[0] = np.NaN
    
shift = 0.05
bars += shift

plt.figure(figsize=[4,8])
colors = ['k', cmap(0), cmap(7), [0.886,0.6745,0.2745]]
types_extra = ['WT (Exp)', 'WT (Sim)', 'Y132F', 'G131D']

for i,t in enumerate(types_extra):
    plt.bar(i,bars[i],capsize=0,color=colors[i],bottom=-shift,fill=(i!=0),linewidth=3)
plt.plot([-0.5,3.5],[0,0],'k--',linewidth=3)

plt.xticks(range(4),types_extra,fontsize=24,rotation=90)
plt.yticks([0,0.2,0.4,0.6,0.8],fontsize=24)

plt.savefig(f'PLCg_frac_clust_bar_plot.png',dpi=1000,bbox_inches='tight')
```
