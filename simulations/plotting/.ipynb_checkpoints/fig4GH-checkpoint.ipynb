```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from glob import glob
import h5py
from matplotlib.ticker import FixedLocator

MHC_density = [round(0.05*(10**(1/3))**i,2) for i in range(9)]
density_array = np.array(MHC_density)
strength = np.array([0.3, 0.03, 0.003])

cmap = plt.get_cmap('Blues')
```

```python
base_dir = '../run_sims/dynamic_range/'
```

```python
max_nclust = pd.read_csv(f'{base_dir}dynamic_range_max_clustered_distr.csv',index_col=0)
TCR_bound = pd.read_csv(f'{base_dir}dynamic_range_n_bound_TCR.csv',index_col=0)

frac_clust = np.zeros([len(strength),len(MHC_density)])
n_bound = np.zeros([len(strength),len(MHC_density)])
for i,stg in enumerate(strength):
    for ii,dnst in enumerate(density_array*(0.3/stg)):
        frac_clust[i,ii] = np.mean(max_nclust[f'str{stg}_den{dnst}']>50)
        n_bound[i,ii] = np.mean(TCR_bound[f'str{stg}_den{dnst}_mean'])

plt.figure(figsize=[10,6])
colors = [cmap(0.99),cmap(0.7),cmap(0.4)]
for i,stg in enumerate(strength):
    plt.semilogx(n_bound[i,:],frac_clust[i,:],'o-',linewidth=3,markersize=10,label=stg*50,color=colors[i])

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.legend(loc=[1.1,0.5],fontsize=24,title='$\\tau _{off}$ (s)',title_fontsize=24)

plt.savefig('frac_clustered_vs_occupancy.png',dpi=1000,bbox_inches='tight')
```

```python
nclust = np.zeros([len(strength),len(MHC_density)])
errors = np.zeros([2,len(strength),len(MHC_density)])
for i,stg in enumerate(strength):
    for ii,dnst in enumerate(density_array*(0.3/stg)):
        active = max_nclust[f'str{stg}_den{dnst}'].values
        active = active[active>0]
        if len(active) == 0:
            nclust[i,ii] = 0
        else:
            nclust[i,ii] = np.mean(active)
            errors[0,i,ii] = np.quantile(active,0.25)
            errors[1,i,ii] = np.quantile(active,0.75)

fig, ax = plt.subplots(figsize=[10,6])
for i in [0,1,2]:
    plt.errorbar(n_bound[i,:],nclust[i,:],yerr=np.abs(errors[:,i,:]-nclust[i,:]),
                 capsize=6,linewidth=3,markersize=10,elinewidth=3,capthick=3,color=colors[i],
                 fmt='o-',label=strength[i]*50,barsabove=True)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.yscale('symlog',linthresh=np.min(nclust[nclust>0]),linscale=0.5)
plt.xscale('log')

minor_locator = FixedLocator(np.concatenate([np.linspace(10,90,9),np.linspace(200,900,8),np.linspace(2000,9000,8)]))
ax.yaxis.set_minor_locator(minor_locator)
ax.set_yticks([0,100,1000,10000])
ax.set_ylim([-10,10000])

plt.legend(loc=[1.1,0.5],fontsize=24,title='$\\tau _{off}$ (s)',title_fontsize=24)

plt.savefig('max_nclust_clustered_vs_occupancy.png',dpi=1000,bbox_inches='tight')
```
