```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from glob import glob
import h5py
from scipy.optimize import curve_fit
```

```python
base_dir = '../run_sims/phase_separation/'
```

```python
density_range = [round(0.006*(2**(i/3))/(0.03**2),2) for i in range(12)]
density_max_nclust = pd.read_csv(f'{base_dir}LAT_only_max_n_clust_distr.csv',index_col=0)
```

## max N clustered

```python
density_nmol = pd.read_csv(f'{base_dir}LAT_only_nmol_distr.csv',index_col=0)

sizes = []
for i,d in enumerate(density_range):
    sizes.append(np.mean(density_max_nclust[f'd{d}']/density_nmol[f'd{d}']))

plt.figure(figsize=[8,8])
plt.plot(np.array(density_range),sizes,'o-',linewidth=3,markersize=10)
plt.semilogx()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

std_limits = (plt.xlim()[0],100)
```

## initiation times

```python
def decay(x,C,e):
    return C - e*x
```

```python
fracs = []
for i,d in enumerate(density_range):
    fracs.append(np.mean(density_max_nclust[f'd{d}']>50))
fracs = np.array(fracs)
```

```python
int_times = pd.read_csv(f'{base_dir}LAT_only_initiation_times.csv',index_col=0)

mean_times = np.zeros([len(density_range)])
errors = np.zeros([2,len(density_range)])
for i,d in enumerate(density_range):
    mean_times[i] = np.nanmean(int_times[f'd{d}']/10)
    errors[0,i] = np.nanquantile(int_times[f'd{d}']/10,0.25)
    errors[1,i] = np.nanquantile(int_times[f'd{d}']/10,0.75)
density_range_array = np.array(density_range)
mean_times_array = np.array(mean_times)

plt.figure(figsize=[6,6])
plt.errorbar(density_range_array[fracs>0.01],mean_times_array[fracs>0.01],yerr=np.abs(errors[:,fracs>0.01]-mean_times[fracs>0.01]),
             capsize=6,linewidth=3,markersize=10,elinewidth=3,capthick=3,fmt='o-',zorder=1)

params,cov = curve_fit(decay,np.log10(density_range_array[fracs>0.5]),np.log10(mean_times_array[fracs>0.5]),
                  p0=[4,3],bounds=[0,np.inf])
x_fit = np.linspace(1.2,2,1000)
y_fit = 10**decay(x_fit,params[0],params[1])
plt.plot(10**x_fit,y_fit,'r-',linewidth=3)

ylims = plt.ylim()
plt.fill_between([std_limits[0],np.min(density_range_array[fracs>0.01])],[2,2],[200,200],color='grey',alpha=0.5)

plt.loglog()
plt.xlim(std_limits)
plt.ylim([2,200])
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

plt.savefig('init_time_vs_density.png',dpi=1000,bbox_inches='tight')
```

## fraction of sims with a cluster

```python
def hill_func(x,K,n):
    return x**n/(K**n + x**n)
```

```python
params,cov = curve_fit(hill_func,density_range_array,fracs,p0=[2e2,2])
x_fit = np.logspace(np.log10(min(density_range_array)),np.log10(max(density_range_array)),1000)
y_fit = hill_func(x_fit,*params)
print(params)

plt.figure(figsize=[6,6])
plt.plot(density_range_array,fracs,'o',linewidth=5,markersize=12)
plt.plot(x_fit,y_fit,'k-',linewidth=3)

plt.semilogx()
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.xlim(std_limits)

plt.savefig('frac_with_clusters_vs_density.png',dpi=1000,bbox_inches='tight')
```
