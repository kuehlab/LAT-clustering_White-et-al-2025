```python
import pickle as pkl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
tab = matplotlib.cm.get_cmap('tab10')
import os

import pandas as pd

from itertools import groupby
import h5py

plt.rcParams['text.usetex'] = False
import pickle as pkl
from scipy.optimize import curve_fit
```

```python
dt = 1.25e-4 #sec
dx = 0.03 #um

bind_weight_0 = 2.5
deg_weight_0 = 2

N_TCR = 2000 #based on BioNUmbers 100/um^2
k_on = 2.5e-5 # based on https://www.science.org/doi/10.1126/scisignal.aat8715 (Jay Groves paper)
               # and BioNumbers: 100 TCR/um^2
               # value calculated for a single TCR, with units of um^2/pMHC per second
k_off = 0.02 #based on MCC from same publication, units are per second
TCR_density = 300 #per um^2 (3x true number so as to have multiple bound at the same time)

box_widths = [1,2,4,7,10,13,16,25,34,49,70,100]
n_MHC = 4

k_create_0 = 120
k_create_1 = 12

N_reps = 500

N_per = 4
N_jobs = int(N_reps/N_per)
print(N_jobs,N_jobs*len(box_widths))
```

```python
with open('cmds.sh','w') as f:
    
    for i,box_width in enumerate(box_widths):
        
        sim_width = box_width + 450
        if sim_width%2 == 0:
            sim_width += 1
        middle = int(np.floor(sim_width/2))
        print(box_width, sim_width, middle)
        
        spacing = (box_width - 1)/3
        
        source_locs = []
        if spacing <= 1:
            factor = int((n_MHC**2)/(box_width**2))
            for x in range(middle - int(np.floor(box_width/2)), middle + int(np.ceil(box_width/2))):
                for y in range(middle - int(np.floor(box_width/2)), middle + int(np.ceil(box_width/2))):
                    for j in range(factor):
                        source_locs.append([x,y])
        
        else:
            spacing = int(spacing)
            for x in range(middle - int((n_MHC-1)*np.ceil(spacing/2)), middle + int((n_MHC-1)*spacing/2) + 1, spacing):
                for y in range(middle - int((n_MHC-1)*np.ceil(spacing/2)), middle + int((n_MHC-1)*spacing/2) + 1, spacing):
                    source_locs.append([x,y])
        
        source_locs = np.array(source_locs)
        print(source_locs.T)
                
        kws = {'init_kws':{'source_locs':source_locs,
                           'k_on_TCR':k_on,
                           'k_off_TCR':k_off,
                           'TCR_density':TCR_density},
               
               'diffusion_kws':{'D_LAT':0.1725, #um^2/s
                                'dx':dx*1000}, #nm

               'update_binding_kws':{'k_bind_max':  {0:3200,            1:3200}, # s^-1
                                     'k_bind_min':  {0:13.2,            1:35.2,},# s^-1
                                     'bind_infl':   {0:6*bind_weight_0, 1:6*bind_weight_0}, #max possible bonds = 30
                                     'k_unbind':    {0:3.66,            1:6.64e-3},# s^-1
                                     'bind_weights':{0:bind_weight_0,   1:bind_weight_0},
                                     'E_first':1, 'E_nth':0},

               'update_exist_kws':{'k_on_TCR':k_on, 'k_off_TCR':k_off,
                                   'TCR_density':TCR_density,
                                   'k_create_max_0':k_create_0, 'k_create_max_1':k_create_1, # s^-1
                                   'deg_weights':   {0:deg_weight_0,          1:4*deg_weight_0},
                                   'k_deactivate':2.8e-2, # s^-1
                                   'k_degrade_min':8e-3, # s^-1
                                   'k_degrade_max':5.3, # s^-1
                                   'degrade_infl':4*deg_weight_0},
               'num_mols':{},
               'cluster_size_distribution':{'max_cluster_size':1500},
               'avg_mol_position':{},
               'num_mols_near_source':{'width':15},
               'n_bound_TCR':{},
               'n_active_TCR':{},
               'n_bound_and_active_TCR':{}
              }
        
        pkl.dump(kws, open(f'kws/width{box_width}_kws.pkl','wb'))

        for j in range(N_jobs):
            f.write('python ../../code/LAT_cluster_sim.py '+\
                    f'-N {N_per} -T 120 -dt {dt} -grid_size {dx*sim_width} -dx {dx} '+\
                    f'-record_every_Nth {int(0.1/dt)} -write_every_Nth 200 -N_save {int(j<5)} '+\
                    '-summary_funcs num_mols cluster_size_distribution '+\
                    'avg_mol_position n_bound_TCR n_active_TCR n_bound_and_active_TCR '+\
                    '-binding_function binding_sigmoid_p_promiscuous '+\
                    '-init_function initialize_blank_pre_eq_with_depletion '+\
                    '-update_exist_function fixed_fast_point_source_with_loss_negfdbk_and_depletion '+\
                    '-diffusion_function diffusion_bessel_push '+\
                    f'-kw_pkl kws/width{box_width}_kws.pkl '+\
                    f'-outfile_base results/width{box_width}_{j}\n')
```

```python
complete = np.zeros([12,N_reps])

for i,w in enumerate(box_widths):
    n = np.zeros([N_reps,1201])
    for j in range(N_jobs):
        try:
            times =  h5py.File(f'results/width{w}_{j}_summary_funcs.h5','r')['time'][:]
            n[j*N_per:(j+1)*N_per,:] = times
        except FileNotFoundError:
            pass

    complete[i,:] = np.all(n[:,1:]>0,axis=1)

complete = complete.astype(bool)
print(N_reps - np.sum(complete,axis=1))
```

```python
to_save = pd.DataFrame()
for i,box_width in enumerate(box_widths):
    plt.figure(figsize=[4,4])

    n = np.zeros([N_reps,1201])
    for j in range(N_jobs):
        n[j*N_per:(j+1)*N_per,:] = h5py.File(f'results/width{box_width}_{j}_summary_funcs.h5',
                                    'r')['n_bound_TCR'][:]
        
    n = n[complete[i],:]

    n_mean = np.mean(n,0)
    n_low = np.quantile(n,0.1,axis=0)
    n_high = np.quantile(n,0.9,axis=0)
    
    to_save[f'width{box_width}_mean'] = n_mean
    to_save[f'width{box_width}_low'] = n_low
    to_save[f'width{box_width}_high'] = n_high

    plt.plot(range(1201),n_mean,label='Total')
    plt.fill_between(range(1201),n_low,n_high,alpha=0.5)

    plt.ylim([-0.03,5])
    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time (sec)')
    plt.ylabel('Number of Bound TCRs')
    plt.title(f'width{box_width}')

to_save['time'] = np.array(range(1201))/10
to_save.to_csv('spacing_n_bound_TCR.csv')
```

```python
to_save = pd.DataFrame()
for i,box_width in enumerate(box_widths):
    plt.figure(figsize=[4,4])
    
    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_jobs):
        size_distr[j*N_per:(j+1)*N_per,:] = h5py.File(f'results/width{box_width}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    cluster_size = np.tile(np.array(range(1,1501)).reshape([1,1500]),[1201,1])
    where_cluster = size_distr > 0
    max_cluster_sizes = np.zeros([size_distr.shape[0]])
    for j in range(size_distr.shape[0]):
        sim_clusters = where_cluster[j,:,:]
        sim_clusters = cluster_size*sim_clusters
        max_cluster_sizes[j] = np.max(np.max(sim_clusters,axis=1),axis=0)

    to_save[f'width{box_width}'] = np.zeros([N_reps])
    to_save.loc[complete[i],f'width{box_width}'] = max_cluster_sizes
    to_save.loc[~complete[i],f'width{box_width}'] = np.nan

    sns.histplot(max_cluster_sizes,bins=range(0,1200,50))

    plt.xlabel('Size of Largest Cluster')
    plt.ylabel('Count')
    plt.title(f'width{box_width}')

to_save.to_csv('spacing_largest_cluster.csv')
```

```python
import os
# os.mkdir('init_time_heatmaps')

for i,box_width in enumerate(box_widths):
    plt.figure(figsize=[4,4])
    
    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_jobs):
        size_distr[j*N_per:(j+1)*N_per,:] = h5py.File(f'results/width{box_width}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    init_heatmap = np.zeros([301,30])
    for s in range(301):
        has_cluster = np.sum(size_distr[:,:,s:],axis=2).astype(bool)
        cluster_locs = np.where(has_cluster)
        init_time = np.zeros([size_distr.shape[0]])
        for time in range(size_distr.shape[0]):
            if np.sum(cluster_locs[0]==time):
                init_time[time] = cluster_locs[1][cluster_locs[0]==time][0]
            else:
                init_time[time] = np.NaN

        init_time_distr,_ = np.histogram(init_time,bins=np.linspace(0,1201,31))
        init_heatmap[s,:] = init_time_distr/np.max(init_time_distr)

    to_save = pd.DataFrame(index=range(1,302),
                           columns=np.linspace(0,1201,31)[:-1]/10,
                           data=init_heatmap)
    to_save.to_csv(f'init_time_heatmaps/width{box_width}.csv')

    sns.heatmap(init_heatmap,vmin=0,vmax=1,cbar=False)
    plt.xticks(range(0,31,6),range(0,121,24))
    plt.yticks(range(0,300,60),range(1,301,60))
    plt.ylim([0,300])
    plt.xlabel('Time to First Cluster (sec)')
    plt.ylabel('Cluster Size Min')
    plt.title(f'width{box_width}')
```
