```python
import pickle as pkl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
tab = matplotlib.cm.get_cmap('tab10')
import os
from glob import glob

import pandas as pd

from itertools import groupby
import h5py

plt.rcParams['text.usetex'] = False
import pickle as pkl
from scipy.optimize import curve_fit
```

```python
dt = 1.25e-4 #sec
dx = 0.03 #um 

density_range = [round(0.006*(2**(i/3))/(dx**2),2) for i in range(12)] #um^-2
print(density_range)

bind_weight_0 = 2.5
deg_weight_0 = 2

N_reps = 250
N_per = np.array([50,50,50,50,25,25,10,5,1,1,1,1]).astype(int)
N_job = (N_reps/N_per).astype(int)
print(N_job*N_per)
```

```python
with open('cmds.sh','w') as f:
    
    for i,density in enumerate(density_range):
                    
        kws = {'init_kws':{'density_0':density,
                           'density_1':0,
                           'source_locs':np.zeros([1,2])}, #fake info, not used in simulaiton
               
               'diffusion_kws':{'D_LAT':0.1725, #um^2/s
                                'dx':dx*1000}, #nm

               'update_binding_kws':{'k_bind_max':  {0:3200,            1:3200}, # s^-1
                                     'k_bind_min':  {0:13.2,            1:35.2,},# s^-1
                                     'bind_infl':   {0:6*bind_weight_0, 1:6*bind_weight_0}, #max possible bonds = 30
                                     'k_unbind':    {0:3.66,            1:6.64e-3},# s^-1
                                     'bind_weights':{0:bind_weight_0,   1:bind_weight_0},
                                     'E_first':1, 'E_nth':0},
               'update_exist_kws':{},
               
               'num_mols':{},
               'cluster_size_distribution':{'max_cluster_size':1500},
              }
        
        pkl.dump(kws, open(f'kws/d{density}_kws.pkl','wb'))

        for j in range(N_job[i]):
            f.write(f'python ../../code/LAT_cluster_sim.py '+
                    f'-N {N_per[i]} -T 120 -dt {dt} -grid_size {dx*151} -dx {dx} '+\
                    f'-record_every_Nth {int(0.1/dt)} -N_save {int(j<5)} '+
                    '-summary_funcs num_mols cluster_size_distribution '+\
                    '-binding_function binding_sigmoid_p_promiscuous '+\
                    '-init_function initialize_random '+\
                    '-update_exist_function create_none '+\
                    '-diffusion_function diffusion_bessel_push '+\
                    f'-kw_pkl kws/d{density}_kws.pkl '+\
                    f'-outfile_base results/d{density}_{j}\n')
```

```python
complete = np.zeros([12,N_reps])

for i,density in enumerate(density_range):
    n = np.zeros([N_reps,1201])
    for j in range(N_job[i]):
        n[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5','r')['time'][:]

    complete[i,:] = n[:,-1]>0

complete = complete.astype(bool)
print(N_reps - np.sum(complete,axis=1))
```

```python
to_save = pd.DataFrame()
to_save['time'] = range(1201)

for i,density in enumerate(density_range):

    plt.figure()
    
    nmol_close = np.zeros([N_reps,1201])
    for j in range(N_job[i]):
        nmol_close[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                    'r')['num_mols'][:]

    nmol_close_mean = np.mean(nmol_close,0)
    nmol_close_low = np.quantile(nmol_close,0.1,axis=0)
    nmol_close_high = np.quantile(nmol_close,0.9,axis=0)
    
    to_save[f'd{density}_mean'] = nmol_close_mean
    to_save[f'd{density}_low'] = nmol_close_low
    to_save[f'd{density}_high'] = nmol_close_high

    plt.plot(range(1201),nmol_close_mean,label='Total')
    plt.fill_between(range(1201),nmol_close_low,nmol_close_high,alpha=0.5)

    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time (sec)')
    plt.ylabel('Num Mol')
    plt.title(f'd{density}')
to_save.to_csv('LAT_only_nmol.csv')
```

```python
for i,density in enumerate(density_range):
    plt.figure(figsize=[4,4])
    
    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    sns.heatmap(np.log10(np.mean(size_distr,axis=0)).T,mask=np.mean(size_distr,axis=0).T==0,vmax=0,cbar=False)

    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time (sec)')
    plt.ylabel('Cluster Size')
    plt.ylim([0,500])
    plt.yticks(range(0,700,70),range(1,701,70))
    plt.title(f'd{density}')
```

```python
to_save = pd.DataFrame()
for i,density in enumerate(density_range):

    plt.figure()

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]

    cluster_size = np.tile(np.array(range(1,1501)).reshape([1,1500]),[1201,1])
    where_cluster = size_distr > 0
    max_cluster_sizes = np.zeros([N_reps])
    for i in range(N_reps):
        sim_clusters = where_cluster[i,:,:]
        sim_clusters = cluster_size*sim_clusters
        max_cluster_sizes[i] = np.max(np.max(sim_clusters,axis=1),axis=0)

    sns.histplot(max_cluster_sizes,bins=np.linspace(0,1501,30))
    to_save[f'd{density}'] = max_cluster_sizes

    plt.xlabel('Size of Largest Cluster')
    plt.ylabel('Count')
    plt.title(f'd{density}')
    
to_save.to_csv('LAT_only_largest_cluster_distr.csv')
```

```python
frac_clust = np.zeros([12])

for i,density in enumerate(density_range):

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]

    has_cluster = np.sum(np.sum(size_distr[:,:,50:],axis=2),axis=1).astype(bool)
    frac_clust[i] = np.mean(has_cluster)

plt.plot(np.array(density_range)/0.03**2,frac_clust,'o-')
plt.semilogx()
plt.xlabel('pLAT/um^2')
plt.ylabel('Fraction with Clusters')
```

```python
plt.figure(figsize=[15,5])

to_save = pd.DataFrame()
for i,density in enumerate(density_range):

    plt.figure()

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    has_cluster = np.sum(size_distr[:,:,50:],axis=2).astype(bool)
    cluster_locs = np.where(has_cluster)
    init_time = np.zeros([size_distr.shape[0]])
    
    for i in range(size_distr.shape[0]):
        if np.sum(cluster_locs[0]==i):
            init_time[i] = cluster_locs[1][cluster_locs[0]==i][0]
        else:
            init_time[i] = np.NaN
            
    to_save[f'd{density}'] = init_time

    sns.histplot(init_time, bins=np.linspace(0,1201,75))
    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time to First Cluster (sec)')
    plt.ylabel('Count')
    plt.title(f'd{density}')
    
to_save.to_csv('LAT_only_initiation_times.csv')
```

```python
to_save = pd.DataFrame()
for i,density in enumerate(density_range):

    plt.figure()

    size_distr = np.zeros([N_reps,1201,1500])
    nmol = np.zeros([N_reps,1201])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
        nmol[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/d{density}_{j}_summary_funcs.h5',
                                     'r')['num_mols'][:]

    frac_clust = np.sum(size_distr[:,:,50:],axis=2)
    max_num_clust = np.max(frac_clust*nmol,axis=1)

    sns.histplot(max_num_clust)
    to_save[f'd{density}'] = max_num_clust

    plt.xlabel('Peak Number Clustered')
    plt.ylabel('Count')
    plt.title(f'd{density}')
    
to_save.to_csv('LAT_only_max_n_clust_distr.csv')
```
