```python
import pickle as pkl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
tab = matplotlib.cm.get_cmap('tab10')
import pandas as pd
import sys

from itertools import groupby
import h5py

plt.rcParams['text.usetex'] = False
import pickle as pkl
from scipy.optimize import curve_fit

sys.path.append('../../code/')
from LAT_cluster_sim import diffusion_rate_Bessel
```

```python
dt_base = 1.25e-4 #sec
dx = 0.03 #um

source_locs = np.array([[225,225]])

bind_weight_0 = 2.5
deg_weight_0 = 2

k_create_0 = [120, 120, 132, 92] #s^-1
k_create_1 = [12,  12,  0,   40] #s^-1
dt_factor  = [1,   1/4, 1,   1]

types = {
    0:'Normal',
    1:'Small dt',
    2:'Y132F',
    3:'G135D'
}

N_reps = 1500
N_per = np.array([15,4,50,5])
N_job = (N_reps/N_per).astype(int)
print(N_job)
```

```python
with open('cmds.sh','w') as f:

    for i in range(4):

        dt = dt_base*dt_factor[i]

        kws = {'init_kws':{'source_locs':source_locs.tolist()},

               'diffusion_kws':{'D_LAT':0.043125, #um^2/s
                                'dx':dx*1000}, #nm

               'update_binding_kws':{'k_bind_max':  {0:3200,            1:3200}, # s^-1
                                     'k_bind_min':  {0:13.2,            1:35.2,},# s^-1
                                     'bind_infl':   {0:6*bind_weight_0, 1:6*bind_weight_0}, #max possible bonds = 30
                                     'k_unbind':    {0:3.66,            1:6.64e-3},# s^-1
                                     'bind_weights':{0:bind_weight_0,   1:bind_weight_0},
                                     'E_first':1, 'E_nth':0},

               'update_exist_kws':{'D_TCR_free':0.064, #um^2/s
                                   'D_TCR_bound':0.0043, #um^2/s
                                   'dx':dx*1000, #nm
                                   'k_on_TCR':np.inf,
                                   'k_off_TCR':0,
                                   'k_create_max_0':k_create_0[i], 'k_create_max_1':k_create_1[i], # s^-1
                                   'deg_weights':   {0:deg_weight_0,          1:4*deg_weight_0},
                                   'k_deactivate':2.8e-2, # s^-1
                                   'k_degrade_min':8e-3, # s^-1
                                   'k_degrade_max':5.3, # s^-1
                                   'degrade_infl':4*deg_weight_0},
               'num_mols':{},
               'cluster_size_distribution':{'max_cluster_size':1500},
               'avg_mol_position':{},
              }
        pkl.dump(kws, open(f'kws/type{i}_kws.pkl','wb'))

        for j in range(N_job[i]):
            f.write(f'python ../../code/LAT_cluster_sim.py '+
                    f'-N {N_per[i]} -T 120 -dt {dt} -grid_size {dx*451} -dx {dx} '+\
                    f'-record_every_Nth {int(0.1/dt)} -N_save {int((j<50)and(i==0))} '+
                    '-summary_funcs num_mols cluster_size_distribution avg_mol_position '+
                    '-binding_function binding_sigmoid_p_promiscuous '+
                    '-init_function initialize_blank_all_bound '+
                    '-update_exist_function diffusing_fast_point_source_with_loss_negfdbk '+
                    '-diffusion_function diffusion_bessel_push '+
                    f'-kw_pkl kws/type{i}_kws.pkl '+
                    f'-outfile_base results/type{i}_{j}\n')
```

```python
complete = np.zeros([4,N_reps])

for i,ty in enumerate(range(4)):
    n = np.zeros([N_reps,1201])
    for j in range(N_job[i]):
        n[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                    'r')['time'][:]

    complete[i,:] = n[:,-1]>0

complete = complete.astype(bool)
print(N_reps - np.sum(complete,axis=1))
```

```python
plt.figure(figsize=[15,4])
for i,ty in enumerate(range(4)):
    plt.subplot(1,4,i+1)

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    sns.heatmap(np.log10(np.mean(size_distr,axis=0)).T,mask=np.mean(size_distr,axis=0).T==0,vmax=0,cbar=False)

    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time (sec)')
    plt.ylabel('Cluster Size')
    plt.ylim([0,500])
    plt.yticks(range(0,700,70),range(1,701,70))
    plt.title(f'type{ty}')
```

```python
to_save = pd.DataFrame()
plt.figure(figsize=[15,4])
for i,ty in enumerate(range(4)):
    plt.subplot(1,4,i+1)

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    cluster_size = np.tile(np.array(range(1,1501)).reshape([1,1500]),[1201,1])
    where_cluster = size_distr > 0
    max_cluster_sizes = np.zeros([size_distr.shape[0]])
    for j in range(size_distr.shape[0]):
        sim_clusters = where_cluster[j,:,:]
        sim_clusters = cluster_size*sim_clusters
        max_cluster_sizes[j] = np.max(np.max(sim_clusters,axis=1),axis=0)

    to_save[f'type{ty}'] = np.zeros([N_reps])
    to_save.loc[complete[i],f'type{ty}'] = max_cluster_sizes
    to_save.loc[~complete[i],f'type{ty}'] = np.nan

    sns.histplot(max_cluster_sizes,bins=range(0,700,50))

    plt.xlabel('Size of Largest Cluster')
    plt.ylabel('Count')
    plt.title(f'type{ty}')

to_save.to_csv('single_TCR_largest_cluster.csv')
```

```python
to_save = pd.DataFrame()
plt.figure(figsize=[15,4])
for i,ty in enumerate(range(4)):
    plt.subplot(1,4,i+1)

    size_distr = np.zeros([N_reps,1201,1500])
    n_mol = np.zeros([N_reps,1201])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
        n_mol[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['num_mols'][:]
    size_distr = size_distr[complete[i],:]
    n_mol = n_mol[complete[i],:]

    n_clust = np.max(np.sum(size_distr[:,:,50:],axis=2)*n_mol,axis=1)

    to_save[f'type{ty}'] = np.zeros([N_reps])
    to_save.loc[complete[i],f'type{ty}'] = n_clust
    to_save.loc[~complete[i],f'type{ty}'] = np.nan

    sns.histplot(n_clust,bins=range(0,700,50))

    plt.xlabel('Size of Largest Cluster')
    plt.ylabel('Count')
    plt.title(f'type{ty}')
to_save.to_csv('single_TCR_max_clustered_distr.csv')
```

```python
to_save = pd.DataFrame()
plt.figure(figsize=[15,4])
for i,ty in enumerate(range(4)):
    plt.subplot(1,4,i+1)

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    has_cluster = np.sum(size_distr[:,:,50:],axis=2).astype(bool)
    cluster_locs = np.where(has_cluster)
    init_time = np.zeros([size_distr.shape[0]])

    for t in range(size_distr.shape[0]):
        if np.sum(cluster_locs[0]==t):
            init_time[t] = cluster_locs[1][cluster_locs[0]==t][0]
        else:
            init_time[t] = np.NaN

    to_save[f'type{ty}'] = np.zeros([N_reps])
    to_save.loc[complete[i],f'type{ty}'] = init_time
    to_save.loc[~complete[i],f'type{ty}'] = np.nan

    sns.histplot(init_time, bins=np.linspace(0,1201,30))
    plt.xticks(range(0,1201,240),np.linspace(0,120,6).astype(int))
    plt.xlabel('Time to First Cluster (sec)')
    plt.ylabel('Count')
    plt.title(f'type{ty}')
to_save.to_csv('single_TCR_init_time.csv')
```

```python
import os
os.mkdir('init_time_heatmaps')

plt.figure(figsize=[15,4])
for i,ty in enumerate(range(4)):
    plt.subplot(1,4,i+1)

    size_distr = np.zeros([N_reps,1201,1500])
    for j in range(N_job[i]):
        size_distr[j*N_per[i]:(j+1)*N_per[i],:] = h5py.File(f'results/type{ty}_{j}_summary_funcs.h5',
                                     'r')['cluster_size_distribution'][:]
    size_distr = size_distr[complete[i],:]

    init_heatmap = np.zeros([301,30])
    for s in range(301):
        has_cluster = np.sum(size_distr[:,:,s:],axis=2).astype(bool)
        cluster_locs = np.where(has_cluster)
        init_time = np.zeros([size_distr.shape[0]])
        for time in range(size_distr.shape[0]):
            if np.sum(cluster_locs[0]==time):
                init_time[time] = cluster_locs[1][cluster_locs[0]==time][0]
            else:
                init_time[time] = np.NaN

        init_time_distr,_ = np.histogram(init_time,bins=np.linspace(0,1201,31))
        init_heatmap[s,:] = init_time_distr/np.max(init_time_distr)

    to_save = pd.DataFrame(index=range(1,302),
                           columns=np.linspace(0,1201,31)[:-1]/10,
                           data=init_heatmap)
    to_save.to_csv(f'init_time_heatmaps/type{ty}.csv')

    sns.heatmap(init_heatmap,vmin=0,vmax=1,cbar=False)
    plt.xticks(range(0,31,6),range(0,121,24))
    plt.yticks(range(0,300,60),range(1,301,60))
    plt.ylim([0,300])
    plt.xlabel('Time to First Cluster (sec)')
    plt.ylabel('Cluster Size Min')
    plt.title(f'type{ty}')
```
