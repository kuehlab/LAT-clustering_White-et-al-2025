```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from scipy.stats import gaussian_kde
from scipy.stats import ttest_ind
from scipy.stats import linregress
from glob import glob
import h5py
```

## example distributions

```python
min_n_slices = 2
max_GFP_med = 10**2.5
min_GFP_tot = 100
min_frac_in_cell = 0.99

coefs = pd.read_csv('20230914_csvs/GFP_coefs.csv',names=[1,2]).values[0,:]
coefs
```

```python
for ratio in ['1-3','0']:
    
    array_data = pd.read_csv(f'20230914_csvs/r{ratio}_5min.csv')
    
    array_data['N_GFP'] = (array_data['GFP_tot'] - coefs[1])/coefs[0]
    
    filtered_data = array_data[(array_data['GFP_med'] < max_GFP_med) & 
                               (array_data['n_slices'] >= min_n_slices) &
                               (array_data['frac_in_cell'] > min_frac_in_cell) &
                               (array_data['N_GFP'] > min_GFP_tot)]

    colors = ['cyan','red']
    lows = [-150,-50]
    highs = [300,150]
    ticks = [np.linspace(lows[0],highs[0],4)/100,np.linspace(lows[1],highs[1],5)/100]
    
    plt.figure(figsize=[20,8.75])
    
    for i,ch in enumerate(['pCD3z','pLAT']):
        
        for ii,q in enumerate([0.33,0.66,1]):
            
            low = np.quantile(filtered_data['N_GFP'],q-0.33)
            high = np.quantile(filtered_data['N_GFP'],q)
            subset = filtered_data[(filtered_data['N_GFP'] >= low) & (filtered_data['N_GFP'] < high)]

            plt.subplot(2,3,3*i + ii + 1)

#             real_signal = (subset[f'{ch}_med'] + subset[f'{ch}_bkgnd'])#/filtered_data['GFP_tot']
            real_signal = subset[f'{ch}_med']
            kde_real = gaussian_kde(real_signal)

#             rand_signal = (subset[f'rand_{ch}_med'] + subset[f'rand_{ch}_bkgnd'])#/filtered_data['GFP_tot']
            rand_signal = subset[f'rand_{ch}_med']
            kde_rand = gaussian_kde(rand_signal[~np.isnan(rand_signal)])

            x = np.linspace(lows[i],highs[i],1000)

            y_rand = kde_rand.evaluate(x)
            y_rand = y_rand/np.max(y_rand)
            plt.plot(x/100,y_rand,'k',label='Bkgnd',linewidth=8)
            plt.fill_between(x/100,y_rand,np.zeros([1000]),alpha=0.5,color='k')

            y_real = kde_real.evaluate(x)
            y_real = y_real/np.max(y_real)
            plt.plot(x/100,y_real,label='Arrays',color=colors[i],linewidth=8)
            plt.fill_between(x/100,y_real,np.zeros([1000]),alpha=0.5,color=colors[i])
            
            if ch =='pLAT':
                rand_high = np.nanquantile(subset[f'rand_{ch}_med'],0.95)
                plt.plot([rand_high/100]*2,[-0.5,1.5],'k--',linewidth=8)
                
                print(ch, ratio, q, np.mean(subset[f'{ch}_med']>rand_high))
            else:
                print(ch, ratio, q, np.mean(subset[f'{ch}_med']) - np.mean(subset[f'rand_{ch}_med']))

            # sns.histplot(real_signal,label='real')
            # sns.histplot(rand_signal,label='rand',color='red')
            # plt.xlim([low,high])
            # if ii == 2:
            #     leg=plt.legend(loc=[1.1,0.5],fontsize=48)
            #     for line in leg.get_lines():
            #         line.set_linewidth(16.0)
                
#             if ch == 'pLAT':
            plt.xticks(ticks[i],fontsize=40)
#             else:
#                 plt.xticks(ticks[i],['','','',''])
                
            if q == 0.33:
                plt.yticks([0,0.33,0.66,1],fontsize=40)
            else:
                plt.yticks([0,0.33,0.66,1],['']*4)
            plt.ylim([-0.05,1.05])
            plt.tight_layout()

    plt.savefig(f'{ratio}_5min_med_kdes.png',dpi=1000,bbox_inches='tight')
```

```python
plt.figure(figsize=[4,4])
plt.plot([0,1],[0,1],'k',label='Bkgnd',linewidth=4)
plt.plot([0,1],[0,1],label='Arrays (pCD3$\zeta$)',linewidth=4,color=colors[0])
plt.plot([0,1],[0,1],label='Arrays (pLAT)',linewidth=4,color=colors[1])
plt.legend(loc=[1.1,0.7],fontsize=12,ncol = 3)
plt.savefig(f'line_legend.png',dpi=1000,bbox_inches='tight')
```

## line plots

```python
quantiles = [0.33,0.66,1]

for ratio in ['1-3','0']:
    
    array_data = pd.read_csv(f'csvs/r{ratio}_5min.csv')
    array_data['rep'] = array_data['fname'].map(lambda x: x[0])
    
    array_data['N_GFP'] = (array_data['GFP_tot'] - coefs[1])/coefs[0]
    
    filtered_data = array_data[(array_data['GFP_med'] < max_GFP_med) & 
                               (array_data['n_slices'] >= min_n_slices) &
                               (array_data['frac_in_cell'] > min_frac_in_cell) &
                               (array_data['N_GFP'] > min_GFP_tot)]

    colors = ['cyan','red']
    lows = [50,50]
    highs = [800,350]
    ticks = [np.linspace(lows[0],highs[0],4),np.linspace(lows[1],highs[1],4)]
    
    plt.figure(figsize=[8,16])
    
    for i,ch in enumerate(['pCD3z','pLAT']):
        
        real_means = np.zeros([3])
        real_err = np.zeros([3])
        rand_means = np.zeros([3])
        rand_err = np.zeros([3])
        for ii,q in enumerate(quantiles):
            
            low = np.quantile(filtered_data['N_GFP'],q-0.33)
            high = np.quantile(filtered_data['N_GFP'],q)
            subset = filtered_data[(filtered_data['N_GFP'] >= low) & (filtered_data['N_GFP'] < high)]
            
            groups = subset.groupby('rep')
            real_group_means = []
            rand_group_means = []
            for g,group in groups:
                real_group_means.append(np.nanmedian(group[f'{ch}_med'] + group[f'{ch}_bkgnd']))
                rand_group_means.append(np.nanmedian(group[f'rand_{ch}_med'] + group[f'rand_{ch}_bkgnd']))
            
            real_means[ii] = np.mean(real_group_means)
            rand_means[ii] = np.mean(rand_group_means)
            
            real_err[ii] = np.std(real_group_means)
            rand_err[ii] = np.std(rand_group_means)
        
        plt.subplot(2,1,i+1)
        
        plt.errorbar(quantiles,rand_means,yerr=rand_err,fmt='o-',color='k',label='Background',
                     capsize=12,linewidth=8,markersize=20,elinewidth=6,capthick=6)
        
        plt.errorbar(quantiles,real_means,yerr=real_err,fmt='o-',color=colors[i],label='Arrays',
                     capsize=12,linewidth=8,markersize=20,elinewidth=6,capthick=6)
        
#         plt.plot(quantiles,real_means,'o-',color=colors[i],label='Arrays',linewidth=4,markersize=10)
#         plt.plot(quantiles,rand_means,'ko-',label='Background',linewidth=4,markersize=10)
        
#         leg=plt.legend(loc=[1.1,0.5],fontsize=46)
#         for line in leg.get_lines():
#             line.set_linewidth(16.0)
        
        if ch == 'pLAT':
            plt.ylim([107,144])
            plt.yticks([110,120,130,140],fontsize=48)
            plt.xticks(quantiles,['low','int','high'],fontsize=48)
        else:
            plt.ylim([195,315])
            plt.yticks([200,225,250,275,300],fontsize=48)
            plt.xticks(quantiles,['','',''],fontsize=48)
        plt.tight_layout()

    plt.savefig(f'{ratio}_5min_med_vs_GFP_lineplot.png',dpi=1000,bbox_inches='tight')
```
